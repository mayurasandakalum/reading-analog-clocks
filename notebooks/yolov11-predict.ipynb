```python
from ultralytics import YOLO
import cv2
import matplotlib.pyplot as plt
```

```python
# Load the YOLOv11 Nano model
model = YOLO("yolo11n.pt")

# Load the input image
image_path = "./../data/test_images/test-clock-5.jpg"  # Replace with your image path
image = cv2.imread(image_path)

# Perform object detection with adjusted parameters
results = model(image, conf=0.5, iou=0.4)  # Adjust conf and iou thresholds as needed

# Display the results using matplotlib
for result in results:
    result_image = result.plot()  # Get the image with predictions
    plt.imshow(cv2.cvtColor(result_image, cv2.COLOR_BGR2RGB))  # Convert BGR to RGB
    plt.axis('off')  # Hide axes
    plt.show()
```

```python
# List of image paths
image_paths = [
    "./../data/test_images/test-clock-1.jpg",
    "./../data/test_images/test-clock-2.jpg",
    "./../data/test_images/test-clock-3.png",
    "./../data/test_images/test-clock-4.png",
    "./../data/test_images/test-clock-5.jpg",
    "./../data/test_images/test-clock-6.jpg"
    "./../data/test_images/test-clock-7.jpg"
]

# Loop through each image path, perform detection, and display results
for image_path in image_paths:
    image = cv2.imread(image_path)
    results = model(image, conf=0.5, iou=0.4)
    for result in results:
        result_image = result.plot()
        plt.imshow(cv2.cvtColor(result_image, cv2.COLOR_BGR2RGB))
        plt.axis('off')
        plt.show()
```
